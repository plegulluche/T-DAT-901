```python
#!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
import pandas as pd

# List fiat and crypto currencies into lists 
fiat = ['EUR','USD']
crypto = [
    'BTC',
        'ETH',
        'USDT',
        'BNB',
        'SOL',
        'XRP',
        'USDC',
        'ADA',
        'AVAX',
        'DOGE',
        'LINK',
        'TRX',
        'DOT',
        'MATIC',
        'TON',
        'ICP',
        'SHIB',
        'DAI',
        'LTC',
        'BCH'
    ]

cross = []

# concatenate fiat and crypto currencies into another list to make crosses
for i in range(len(fiat)):
    for y in range(len(crypto)):
        cross.append(crypto[y] + '-' + fiat[i])

cross
```

```python
class Crypto:
    def __init__(self,start_date,end_date,cross):
        self.start_date = start_date
        self.end_date = end_date
        self.cross = cross

    def fetch_crypto(self,start_date,end_date,cross):
        #Creation of an empty template df to get the columns name
        data = yf.download("BTC-USD", start="2020-01-01", end="2020-01-02")
        data = data.reset_index().drop(0)
        data['Cross'] = ''
        df_raw = data
        for i in range(len(cross)):
            data = yf.download(self.cross[i], start=self.start_date, end=self.end_date)
            data = data.reset_index()
            data['Cross'] = cross[i]
            df_raw = pd.concat([df_raw,data])

        return df_raw


start_date="2020-01-01"
end_date="2024-01-01"
data = Crypto(start_date,end_date,cross)
```

```python
df =data.fetch_crypto(start_date,end_date,cross)
```
